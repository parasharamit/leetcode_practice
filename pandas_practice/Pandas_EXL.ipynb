```python
edf = pd.DataFrame({"job":["sales", "sales", "sales", "sales", "sales",
                           "market", "market", "market", "market", "market"],
                    "source":["A", "B", "C", "D", "E", "A", "B", "C", "D", "E"],
                    "count":[2, 4,6,3,7,5,3,2,4,1],
                    "other_col":[1,2,3,4,56,6,3,4,6,11]})
```

```python
edf
```

```python
edf.groupby(["job", "source"]).agg({"count": "sum"})
```

```python
edf.groupby(["job", "source"]).agg({"count": "sum"})
```

```python
edf.groupby(["job", "source"], group_keys=True).apply(lambda x: x.sort_values(["job", "source"], ascending=[False, False]).head(2))
```

```python
 edf.sort_values(['job','count'],ascending=False).groupby(['job', 'source']).head(3)
```

```python
df_agg = edf.groupby(['job','source']).agg({'count':sum})
df_agg
```

```python
g = df_agg['count'].groupby('job', group_keys=False)
g.nlargest(3)
```

```python
edf.groupby(['job','source'])['count'].sum().groupby('job', group_keys=False).nlargest(3)
```
