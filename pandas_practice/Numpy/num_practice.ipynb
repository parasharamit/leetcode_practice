```python
import numpy as np 
from keras.models import Sequential 
from keras.layers import AveragePooling2D 
 
image = np.array([[8.0, 2.0, 6.0, 9.0], 
         [6.0, 4.0, 1.0, 4.0], 
         [1.0, 6.0, 4.0, 1.0], 
         [4.0, 9.0, 5.0, 6.0]]) 
image = image.reshape(1, 4, 4, 1) 
 
model = Sequential( 
  [AveragePooling2D(pool_size = 2, strides = 2)]) 
 
out = model.predict(image) 
 
out = np.squeeze(out) 

print(out) 
```
